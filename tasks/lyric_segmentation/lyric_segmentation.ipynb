```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def read_lyrics_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        documents = [line.strip() for line in file if line.strip()]
    return documents

def similarity(s1,s2):
  tfidf = TfidfVectorizer().fit_transform([s1, s2])
  return cosine_similarity(tfidf[0:1], tfidf[1:2])[0][0]


# TODO: This method cannot make pre-chorus and chorus into 2 segments
def find_longest_similar_sections(documents, threshold=0.5):
    longest_section = []
    longest_section_length = 0  # 记录最长段落的行数

    for i in range(len(documents)):
        for j in range(i+1, len(documents)):
            k = 0
            while i + k < len(documents) and j + k < len(documents):
                section_1 = ' '.join(documents[i:i + k + 1])
                section_2 = ' '.join(documents[j:j + k + 1])
                if similarity(section_1, section_2) > threshold:
                    print(documents[i:k+1])
                    k += 1
                else:
                    break

            # 如果当前找到的重复段落比之前的都长，则更新最长段落
            if k > longest_section_length:
                longest_section_length = k
                longest_section = documents[i:i + k]  # 记录最长段落的内容

    return longest_section
```

```python
file_path = '/Users/jie/dev/DBMS/jie_test_lyric/Bitch_Dont_Kill_My_Vibe.txt'
documents = read_lyrics_from_file(file_path)


print(len(documents))
print(documents[0:5])

longest_repeat_section_content = find_longest_similar_sections(documents, threshold=0.9)
print(longest_repeat_section_content)
```
